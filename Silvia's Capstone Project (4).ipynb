### Building the Dataframe

```python
data = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
df=data[0]
df.head()
```

### Data preparation

```python
df1=df.replace('Not assigned',np.NaN)
```

```python
df.shape
```

```python
df=df1.dropna(subset=["Borough"], axis=0)
df.head(10)
```

```python
df1= df.groupby(by=['Postal Code','Borough']).agg(lambda x: ','.join(x))
```

```python
df1.reset_index(level=['Postal Code','Borough'], inplace=True) 
```

```python
df1.rename(columns={'Postal Code': 'PostalCode'}, inplace=True)
```

```python
df1.head(20)
```

```python
df=df1
```

```python
df["Neighbourhood"].replace(np.nan, 'Borough', inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df_latlong = pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
df_latlong.head()
```

```python
df_latlong.rename(columns={'Postal Code': 'PostalCode'}, inplace=True)
df_latlong.head()
```

```python
df = df.merge(df_latlong, left_on='PostalCode', right_on='PostalCode')
```
